```python
app_med = app.query('event == "purchase"') \
                .drop_duplicates('device_id') \
                .groupby('utm_source') \
                .agg({'purchase_sum' : 'median'}) \
                .sort_values(by='purchase_sum', ascending=False)
app_med
```
