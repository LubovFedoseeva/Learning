```python
taxi = pd.read_csv('/home/jupyter-l.fedoseeva-12/Lesson_4/taxi_peru (1).csv', sep=';')
```

```python
taxi.head()
```

```python
taxi.shape
```

```python
taxi['start_at'] = pd.to_datetime(taxi['start_at'])
taxi['end_at'] = pd.to_datetime(taxi['end_at'])
taxi['arrived_at'] = pd.to_datetime(taxi['arrived_at'])
```

```python
taxi[['start_at', 'end_at', 'arrived_at']].dtypes
```

```python
taxi = taxi.query('start_type == "asap" or start_type == "reserved" ')
```

```python
taxi.shape
```

```python
taxi
```

```python
taxi.arrived_at - taxi.start_at
```

```python
taxi['wait_time'] = (taxi.arrived_at - taxi.start_at).astype('timedelta64[m]')
```

```python
(taxi.arrived_at - taxi.start_at).astype('timedelta64[m]')
```

```python
new_columns = taxi.arrived_at - taxi.start_at
taxi['wait_time'] = new_columns.astype('timedelta64[m]')
```

```python
taxi.wait_time.dtypes
```

```python
taxi
```

```python
count_late = taxi.query('start_type == "reserved" and  wait_time > 0.0') \
    .groupby('driver_id', as_index=False) \
    .agg({'user_id': 'count'}) \
    .rename(columns={'user_id': 'count_late'}) \
    .sort_values('count_late', ascending=False)
```

```python
count_late
```

```python
taxi.query('start_type == "reserved" and  wait_time > 0.0') \
    .groupby('driver_id') \
    .size() \
    .sort_values(ascending=False)

# второй способ
```

```python
taxi['month'] = taxi['start_at'].dt.month
```
